```python
import numpy as np
import matplotlib.pyplot as plt
import math

import string, unicodedata
from pyspark.sql import functions as F

def fingerprint(value):
    key = unicodedata.normalize('NFKD', value).encode('ascii','ignore').decode()
    key = set(key.strip().lower().translate(str.maketrans('','',string.punctuation)).split())
    key = ' '.join(sorted(list(key)))
    return (key, value)

def fingerprint_lenient(value):
    key = unicodedata.normalize('NFKD', value).encode('ascii','ignore').decode()
    key = list(key.strip().lower().translate(str.maketrans('','',string.punctuation)).split())
    key = ' '.join(sorted(list(key)))
    return (key, value)
```

```python
restaurant_rdd = inspection_df.select('DBA').rdd.flatMap(list)
print(restaurant_rdd.take(5))
duplicates = restaurant_rdd.distinct(). \
	map(fingerprint_lenient). \
	groupByKey(). \
	filter(lambda x: len(x[1])>1). \
	mapValues(list). \
	collect()

print(duplicates)
```

```python
replacements = {}
for t in duplicates:
    n = len(t[1])
    for i in range(1, n):
        # mark as a replacement if they have the same cuisine type
        df1 = inspection_df.filter((inspection_df['DBA'] == t[1][i]) | (inspection_df['DBA'] == t[1][0]))
        if df1.select('CUISINE DESCRIPTION').distinct().count() == 1:
            replacements[t[1][i]] = t[1][0]
```

```python
replacements
```

```python
inspection_df = inspection_df.replace(replacements, subset = 'DBA')
```

```python
cuisine_rdd = inspection_df.select('CUISINE DESCRIPTION').rdd.flatMap(list)
print(cuisine_rdd.take(5))
duplicates = cuisine_rdd.distinct(). \
	map(fingerprint). \
	groupByKey(). \
	filter(lambda x: len(x[1])>1). \
	mapValues(list). \
	collect()

print(duplicates)
```

```python
# no replacement required as all cuisines have a different fingerprint key
```

```python

it_rdd = inspection_df.select('INSPECTION TYPE').rdd.flatMap(list)
print(it_rdd.take(5))
duplicates = it_rdd.distinct(). \
	map(fingerprint). \
	groupByKey(). \
	filter(lambda x: len(x[1])>1). \
	mapValues(list). \
	collect()

print(duplicates)
```

```python
#See which violations were isssued most in 2019 and 2022 to asain and chinese restaurants

inspection_df.createOrReplaceTempView("inspections")

Chinese_violations = spark.sql("""
    SELECT t.year, t.cuisine, t.`VIOLATION DESCRIPTION`, t.count
    FROM (
        SELECT YEAR(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy')) as year,
               CASE WHEN `CUISINE DESCRIPTION` IN ('Asian/Asian Fusion', 'Chinese') THEN 'Asian/Chinese' ELSE `CUISINE DESCRIPTION` END as cuisine, 
               `VIOLATION DESCRIPTION`, COUNT(*) as count,
               ROW_NUMBER() OVER (PARTITION BY YEAR(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy')), CASE WHEN `CUISINE DESCRIPTION` IN ('Asian/Asian Fusion', 'Chinese') THEN 'Asian/Chinese' ELSE `CUISINE DESCRIPTION` END ORDER BY COUNT(*) DESC) as row_num
        FROM inspections
        WHERE YEAR(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy')) IN (2019, 2022)
              AND (CASE WHEN `CUISINE DESCRIPTION` IN ('Asian/Asian Fusion', 'Chinese') THEN 'Asian/Chinese' ELSE `CUISINE DESCRIPTION` END) IN ('Asian/Chinese')
        GROUP BY YEAR(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy')), `VIOLATION DESCRIPTION`, cuisine
    ) t
    WHERE t.row_num <= 3
    ORDER BY t.year DESC, t.count DESC
""")

# convert the PySpark DataFrame to a pandas DataFrame
Chinese_violations_pd = Chinese_violations.toPandas()


# display the pandas DataFrame
display(Chinese_violations_pd)
```

```python
#Run a SQL query to filter the inspection data by cuisine and date range
inspection_df.createOrReplaceTempView("inspection_data")


results = spark.sql("""
    SELECT `CUISINE DESCRIPTION` as cuisine,
           count(*) as violation_count, YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) as year
    FROM inspection_data
    WHERE MONTH(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy-MM-dd')) BETWEEN 01 AND 05
        AND `CUISINE DESCRIPTION` IN ('Asian', 'Chinese')
    GROUP BY cuisine, year
    HAVING year BETWEEN 2019 AND 2023
    ORDER BY year DESC, cuisine ASC
""")

# Display the results
results.show()
```

```python
#INSPECTIONS INCREASED BY APPROX. 203 PERCENT FROM 2019 TO 2022
inspection_df.createOrReplaceTempView("inspections")
inspection_by_year = spark.sql("""
    SELECT YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) as year, COUNT(*) as count
    FROM inspections
    GROUP BY year
    ORDER BY year ASC
    
""")
inspection_by_year.show()
```

```python
#Ameircan restaruants inreased up by 133 percent from 2019 to 2022
#Chinese restaruants inreased up by 231 percent from 2019 to 2022


#find avg number of restaurants
Asians_restaurants = spark.sql("""
    SELECT YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) as year, 
           CASE WHEN `CUISINE DESCRIPTION` IN ('Asian/Asian Fusion', 'Chinese') THEN 'Asian/Chinese' ELSE `CUISINE DESCRIPTION` END as cuisine,
           COUNT(*) as count
    FROM inspections
    WHERE YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) = 2019 OR YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) = 2022
    GROUP BY year, cuisine
    HAVING cuisine = 'Asian/Chinese' OR cuisine ='American'
    ORDER BY year ASC
""")
Asians_restaurants.show()
```

```python
grades_by_cusine = spark.sql("""
    SELECT YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) as year, grade,
           CASE WHEN `CUISINE DESCRIPTION` IN ('Asian/Asian Fusion', 'Chinese') THEN 'Asian/Chinese' ELSE `CUISINE DESCRIPTION` END as cuisine,
           COUNT(*) as count
    FROM inspections
    WHERE YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) = 2019 OR YEAR(DATE_FORMAT(TO_DATE(`INSPECTION DATE`, 'MM/dd/yyyy'), 'yyyy')) = 2022
    GROUP BY year, grade, cuisine
    HAVING (cuisine = 'Asian/Chinese' OR cuisine ='American') AND grade = 'A'
    ORDER BY grade DESC
""")
grades_by_cusine.show(100)

```
